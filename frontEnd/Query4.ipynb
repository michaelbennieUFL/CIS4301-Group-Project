## Investigating the Impact of Traffic Conditions on Road Safety Outcomes
In the realm of road safety research, understanding the influence of traffic conditions on various accident outcomes is paramount. This investigation delves into three key correlations: the relationship between accident severity, total number of crashes, distance affected by a crash, and the number of traffic conditions present at the time of incidents. Through dynamic plots that allow users to manipulate time periods and observe trends across different levels of traffic congestion, this analysis aims to unravel the complex interplay between traffic conditions and road safety. By examining these correlations, we seek to provide insights that inform targeted interventions, infrastructure improvements, and policy measures aimed at mitigating road hazards and enhancing overall safety on roadways.

```python
labels = ["Zero Conditions", "One to Four Conditions", "Five to Eight Conditions", "Nine or More Conditions"]
Query4SeverityPlot(query4_data, labels, "Query 4: Correlation of Accident Severity and Number of Traffic Conditions")
```

The plot depicts the correlation between accident severity and the number of traffic conditions present at the time of the incident, offering users a dynamic tool to explore the impact of various traffic scenarios on road safety outcomes. With the ability to drag and select specific time periods, users can analyze temporal variations in this relationship, facilitating the identification of trends and patterns over time. The inclusion of four distinct lines representing different levels of traffic conditionsranging from no conditions to 9 or moreenhances the graph's utility by allowing users to compare accident severity trends across different traffic scenarios. 

During the period from January 2016 to January 2019, the plot illustrates distinct trends in the correlation between accident severity and the number of traffic conditions. Notably, the line representing accidents occurring under no traffic conditions shows a consistent downward trend between January 2016 and May 2016, indicating a decrease in accident severity over this time frame. Conversely, the line representing accidents occurring under 1-4 traffic conditions exhibits less volatility and tends towards a flat trajectory, suggesting a relatively stable accident severity pattern within this range of traffic conditions. In contrast, the line representing accidents occurring under 5-8 traffic conditions displays greater volatility, with fluctuations observed over time. Notably, throughout this period, the line representing accidents under no traffic conditions consistently remains above those for 1-4 and 5-8 conditions, indicating that accidents occurring under less congested traffic conditions tend to have higher severity levels compared to those occurring under moderate to heavy traffic conditions. 

Researching the relationship between accident severity and the number of traffic conditions is vital for several reasons. Firstly, it provides insights into the role of traffic congestion, road infrastructure, and other conditions in contributing to accidents of varying severity. Secondly, understanding how accident severity varies with the number of traffic conditions can inform traffic management strategies, infrastructure improvements, and policy interventions aimed at mitigating road safety hazards and reducing the incidence of severe accidents. Ultimately, such research can contribute to the development of more effective measures to enhance road safety and save lives.
